```python
import pandas as pd
import torch
import pickle
from sklearn.decomposition import PCA
```

```python
file = open('text_embed_final.pickle', 'rb')
data = pickle.load(file)
file.close()
```

```python
text_pca = {}
for scene in list(data.keys()):
    text_embed = data[scene]
    pca_text = PCA(n_components=18)
    text_data = pca_text.fit_transform(text_embed.transpose())
    reduced_embed = text_data.transpose()
#     print(reduced_embed.shape)
    text_pca[scene] = reduced_embed
```

```python
with open('text_embed_pca_final.pickle', 'wb') as handle:
    pickle.dump(text_pca, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
